# [dbo].[depositar]

```sql
/****** Object:  StoredProcedure [dbo].[depositar]    Script Date: 6/06/2023 3:07:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[depositar](@id_cuenta_origen varchar(20), @monto bigint, @id_origen int, @id_tipo int)
AS
DECLARE @error_message NVARCHAR(4000), @error_number INT;
BEGIN
    BEGIN TRY
        BEGIN TRANSACTION;

        -- Verificar que la cuenta exista
        IF EXISTS (SELECT 1 FROM Cuentas_Bancarias WHERE Numero_Cuenta = @id_cuenta_origen) 
        BEGIN
            -- Sumar el monto a la cuenta de origen
            UPDATE Cuentas_Bancarias
            SET Saldo = Saldo + @monto
            WHERE Numero_Cuenta = @id_cuenta_origen;

            -- Insertar en la tabla Transacciones
            INSERT INTO Transacciones (fecha, monto, id_cuenta_origen, id_tipo_transaccion)
            VALUES (GETDATE(), @monto, @id_origen, @id_tipo);

            COMMIT;
        END
        ELSE
        BEGIN
            -- La cuenta no existe
            THROW 50001, 'La cuenta no existe.', 1;
        END;
    END TRY
    BEGIN CATCH
        IF @@TRANCOUNT > 0
            ROLLBACK;

        -- Registrar la transacción fallida en la tabla Transacciones_Fallidas
        INSERT INTO transacciones_fallidas (fecha, tipo_error, cuenta_origen) 
        VALUES (GETDATE(), ERROR_MESSAGE(), @id_cuenta_origen);

        -- Mostrar el error 
        SET @error_message = ERROR_MESSAGE();
        SET @error_number = ERROR_NUMBER();
    END CATCH;

    SELECT @error_message, @error_number;
END;
GO
```

# [dbo].[retirar]

```sql
/****** Object:  StoredProcedure [dbo].[retirar]    Script Date: 6/06/2023 3:07:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE procedure [dbo].[retirar](@id_cuenta_origen varchar(20), @monto bigint, @id_origen int, 
 @id_tipo int)
as
DECLARE @error_message NVARCHAR(4000), @error_number INT;
BEGIN
        BEGIN TRY
            BEGIN TRANSACTION;

            -- Verificar que los usuarios existan
            IF EXISTS (SELECT 1 FROM Cuentas_Bancarias WHERE Numero_Cuenta = @id_cuenta_origen) 
              
            BEGIN
                -- Verificar que el saldo sea mayor o igual al monto enviado
                IF (SELECT saldo FROM Cuentas_Bancarias WHERE Numero_Cuenta = @id_cuenta_origen) >= @monto
                BEGIN
                    -- Restar el monto de la cuenta de origen
                    UPDATE Cuentas_Bancarias
                    SET Saldo = Saldo - @monto
                    WHERE Numero_Cuenta = @id_cuenta_origen;

                

                    -- Insertar en la tabla Transacciones
                    INSERT INTO Transacciones (fecha, monto, id_cuenta_origen, id_tipo_transaccion)
                    VALUES (GETDATE(), @monto, @id_origen,@id_tipo);

                    COMMIT;
                END
                ELSE
                BEGIN
                    -- El saldo es insuficiente
                    THROW 50001, 'El saldo de la cuenta de origen es insuficiente.', 1;
                END;
            END
            ELSE
            BEGIN
                -- Uno o ambos usuarios no existen
                THROW 50002, 'Uno o ambos usuarios no existen.', 1;
            END;
        END TRY
        BEGIN CATCH
            IF @@TRANCOUNT > 0
                ROLLBACK;

            -- Registrar la transacción fallida en la tabla Transacciones_Fallidas
			
            INSERT INTO transacciones_fallidas (fecha, tipo_error, cuenta_origen) 
            VALUES (GETDATE(), ERROR_MESSAGE(), @id_cuenta_origen);

			--mostrar el error 
             SET @error_message = ERROR_MESSAGE();
   			 SET @error_number = ERROR_NUMBER();
            
        END CATCH;
    END;

	select @error_message,@error_number;

	
GO
```

# [dbo].[transferencia]

```sql
/****** Object:  StoredProcedure [dbo].[transferencia]    Script Date: 6/06/2023 3:07:37 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
create procedure [dbo].[transferencia](@id_cuenta_origen varchar(20), @id_cuenta_destino varchar(20), @monto bigint, @id_origen int, 
@id_destino int, @id_tipo int)
as
DECLARE @error_message NVARCHAR(4000), @error_number INT;
	BEGIN
        BEGIN TRY
            BEGIN TRANSACTION;

            -- Verificar que los usuarios existan
            IF EXISTS (SELECT 1 FROM Cuentas_Bancarias WHERE Numero_Cuenta = @id_cuenta_origen) 
                AND EXISTS (SELECT 1 FROM Cuentas_Bancarias WHERE Numero_Cuenta = @id_cuenta_destino)
            BEGIN
                -- Verificar que el saldo sea mayor o igual al monto enviado
                IF (SELECT saldo FROM Cuentas_Bancarias WHERE Numero_Cuenta = @id_cuenta_origen) >= @monto
                BEGIN
                    -- Restar el monto de la cuenta de origen
                    UPDATE Cuentas_Bancarias
                    SET Saldo = Saldo - @monto
                    WHERE Numero_Cuenta = @id_cuenta_origen;

                    -- Sumar el monto a la cuenta de destino
                    UPDATE Cuentas_Bancarias
                    SET Saldo = Saldo + @monto
                    WHERE Numero_Cuenta = @id_cuenta_destino;

                    -- Insertar en la tabla Transacciones
                    INSERT INTO Transacciones (fecha, monto, id_cuenta_origen, id_cuenta_destino, id_tipo_transaccion)
                    VALUES (GETDATE(), @monto, @id_origen, @id_destino, @id_tipo);

                    COMMIT;
                END
                ELSE
                BEGIN
                    -- El saldo es insuficiente
                    THROW 50001, 'El saldo de la cuenta de origen es insuficiente.', 1;
                END;
            END
            ELSE
            BEGIN
                -- Uno o ambos usuarios no existen
                THROW 50002, 'Uno o ambos usuarios no existen.', 1;
            END;
        END TRY
        BEGIN CATCH
            IF @@TRANCOUNT > 0
                ROLLBACK;

            -- Registrar la transacción fallida en la tabla Transacciones_Fallidas
            INSERT INTO transacciones_fallidas (fecha, tipo_error, cuenta_origen, cuenta_destino) 
            VALUES (GETDATE(), ERROR_MESSAGE(), @id_cuenta_origen, @id_cuenta_destino);

            -- Manejar el error como desees
            -- Puedes mostrar un mensaje, registrar el error en una tabla de registro, etc.
            
			 SET @error_message = ERROR_MESSAGE();
   			 SET @error_number = ERROR_NUMBER();
        END CATCH;
    END;  
	
	select @error_message,@error_number;
GO
```
